```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

# Face Classifier

```python
face_classifier = cv2.CascadeClassifier("/content/* Use (plot_img_bef({image-path})) function to plot image before making blur
.xml")
```

# Detect Face

### import image

```python
# import image and convert colors (BGR) to (RGB)
def import_img(path):
  image_src = cv2.imread(path)
  image = image_src[:,:,::-1]  # (BGR) to (RGB)
  image = np.ascontiguousarray(image, dtype=np.uint8)
  return image
```

### Detect and Blur Faces

```python
# Detect and blurring image function
def detect_face(img, scalefactor, minneighbors):
    face_rects=face_classifier.detectMultiScale(img, scalefactor, minneighbors)
    for (x,y,w,h) in face_rects:
        cv2.rectangle(img,
                     (x,y),
                     (x+w,y+h),
                     (255,255,255),
                     0)
        sub_face = img[y:y+h, x:x+w]
        sub_face = cv2.GaussianBlur(sub_face,(17, 17), 30)   #blurring faces
        img[y:y+sub_face.shape[0], x:x+sub_face.shape[1]] = sub_face
    return img    
```

# Test image classifier

```python
#Print image before detect and blur
def plot_img_bef(img_path):
  image = import_img(img_path)   # import image
  plt.title('Image before detect and blur')
  plt.imshow(image)

#Print image after detect and blur
def plot_img_aft(img_path):
  image = import_img(img_path)
  result_img = detect_face(image.copy(), 1.2, 5)
  plt.title('Image after detect and blur')
  plt.imshow(result_img)
```
